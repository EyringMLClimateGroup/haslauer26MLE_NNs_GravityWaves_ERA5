```python
# Plot function

plt.rcParams.update({'font.size': 29})

lonlatbox = None
cmap = cmaps.vik
cbar = 'r'
axis = None
scaling = None
clabel = 'mPa'
projection = None
savefig=False
savefigname=''
titlesize=32
fs2=29


def fourplot_1(plot00, plot01, plot10, plot11, title1='', level=0, title2='', title3='', title4=''):
    i=level

    limits = limits_dict[i]*1000
    bounds = np.linspace(-limits,limits,1000)

    fig = plt.figure(figsize=(30,14.6), dpi=150)

    width, height = 0.48, 0.425
    pad=18.5


    ax00 = fig.add_axes([0.0365,  0.535,  width, height], projection=ccrs.PlateCarree())
    ax01 = fig.add_axes([0.54, 0.535,  width, height], projection=ccrs.PlateCarree())
    ax10 = fig.add_axes([0.0365,  0.025, width, height], projection=ccrs.PlateCarree())
    ax11 = fig.add_axes([0.54, 0.025, width, height], projection=ccrs.PlateCarree())

    psy.plot.mapplot(plot00, ax=ax00, cmap=cmap, title=title1+levels[i]+'', cbar=cbar, clabel=clabel, bounds=bounds,
                     titlesize=titlesize, grid_labelsize=fs2, cticks=np.linspace(-limits, limits, 11), cticksize=fs2, clabelsize=fs2, extend='both')
    ax00.set_title(title1+levels[i], pad=pad, fontsize=titlesize)

    psy.plot.mapplot(plot01, ax=ax01, cmap=cmap, title=title2+levels[i]+'', cbar=cbar, clabel=clabel, bounds=bounds,
                     titlesize=titlesize, grid_labelsize=fs2, cticks=np.linspace(-limits, limits, 11), cticksize=fs2, clabelsize=fs2, extend='both')
    ax01.set_title(title2+levels[i], pad=pad, fontsize=titlesize)

    psy.plot.mapplot(plot10, ax=ax10, cmap=cmap, title=title3+levels[i]+'\n', cbar=cbar, clabel=clabel, bounds=bounds,
                     titlesize=titlesize, grid_labelsize=fs2, cticks=np.linspace(-limits, limits, 11), cticksize=fs2, clabelsize=fs2, extend='both')
    ax10.set_title(title3+levels[i], pad=pad, fontsize=titlesize)

    psy.plot.mapplot(plot11, ax=ax11, cmap=cmap, title=title4+levels[i]+'\n', cbar=cbar, clabel=clabel, bounds=bounds,
                     titlesize=titlesize, grid_labelsize=fs2, cticks=np.linspace(-limits, limits, 11), cticksize=fs2, clabelsize=fs2, extend='both')
    ax11.set_title(title4+levels[i], pad=pad, fontsize=titlesize)

    #fig.savefig("/p/project1/icon-a-ml/haslauer1/____project1/_publication_4500/_20260608_GWMF_MIX_lev" + str(i).zfill(2) + ".png", dpi=150)
```
